```python
import requests
import json
import pandas as pd
from datetime import datetime
from IPython.display import display

# Your RapidAPI Key
API_KEY = "f5056e9848mshb75353066d979e2p1c9df6jsn8bfadc0de1f5"

# API URL to get MotoGP news
API_URL = "https://motogp-news.p.rapidapi.com/motogpnews/0/0"

# Headers for authentication
HEADERS = {
    "x-rapidapi-host": "motogp-news.p.rapidapi.com",
    "x-rapidapi-key": API_KEY
}

# Function to fetch MotoGP news
def fetch_motogp_news():
    response = requests.get(API_URL, headers=HEADERS)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

# Function to extract the first 3 news items
def extract_first_three_news(news_data):
    if isinstance(news_data, str):
        news_data = json.loads(news_data)

    if "MotoGPNews" in news_data and len(news_data["MotoGPNews"]) >= 3:
        articles = news_data["MotoGPNews"][:3]  # Extract first 3 articles
        extracted_articles = []

        for article in articles:
            extracted_articles.append({
                "Title": article.get("title", "No Title"),
                "Link": article.get("link", ""),
                "Description": article.get("description", ""),
                "Category": article.get("category", ""),
                "Published Date": article.get("pubDate", ""),
                "Image URL": article.get("imageURL", "")
            })

        return pd.DataFrame(extracted_articles)  # Convert to DataFrame
    else:
        print("Less than 3 news articles found.")
        return pd.DataFrame()

# Function to generate a stylish HTML newsletter
def generate_newsletter_html(news_df):
    if news_df.empty:
        return "<h3>No news available for today.</h3>"

    html_content = f"""
    <!DOCTYPE html>
    <html>
    <head>
        <style>
            @import url('https://fonts.googleapis.com/css2?family=Poppins:wght@300;400;600&display=swap');

            body {{
                font-family: 'Poppins', sans-serif;
                background-color: #121212;
                color: #ffffff;
                padding: 20px;
                text-align: center;
            }}
            h1 {{
                color: #ff3b30;
                font-size: 36px;
                font-weight: bold;
                text-transform: uppercase;
                letter-spacing: 2px;
                margin-bottom: 15px;
            }}
            .newsletter-container {{
                background-color: #222;
                border-radius: 12px;
                padding: 25px;
                max-width: 800px;
                margin: auto;
                box-shadow: 0px 6px 12px rgba(255, 255, 255, 0.2);
                text-align: center;
            }}
            .intro-text {{
                font-size: 18px;
                line-height: 1.8;
                padding: 15px;
                text-align: justify;
                font-style: italic;
                color: #dddddd;
                background: #1a1a1a;
                padding: 10px;
                border-radius: 10px;
            }}
            .news-box {{
                border: 2px solid #ff3b30;
                padding: 20px;
                border-radius: 8px;
                margin-bottom: 20px;
                background: #1a1a1a;
                transition: transform 0.3s, box-shadow 0.3s;
            }}
            .news-box:hover {{
                transform: scale(1.03);
                box-shadow: 0px 6px 12px rgba(255, 255, 255, 0.3);
            }}
            .news-title {{
                font-size: 22px;
                font-weight: bold;
                color: #ffcc00;
                margin-bottom: 10px;
            }}
            .news-desc {{
                font-size: 16px;
                line-height: 1.8;
                padding: 10px 20px;
                text-align: justify;
                color: #eeeeee;
            }}
            .news-image {{
                width: 100%;
                max-width: 750px;
                display: block;
                margin: auto;
                border-radius: 12px;
                box-shadow: 0px 4px 8px rgba(255, 255, 255, 0.2);
            }}
            .news-link {{
                display: inline-block;
                font-size: 18px;
                font-weight: bold;
                color: #ffffff !important;
                text-decoration: none;
                margin-top: 15px;
                padding: 12px 22px;
                background-color: #007bff;
                border-radius: 8px;
                transition: 0.3s;
            }}
            .news-link:hover {{
                background-color: #0056b3;
                color: #ffffff !important;
            }}
            .footer {{
                margin-top: 20px;
                font-size: 14px;
                color: #bbbbbb;
                font-style: italic;
            }}
            .credits {{
                font-size: 14px;
                color: #ffcc00;
                text-align: center;
                margin-top: 30px;
                font-weight: bold;
            }}
        </style>
    </head>
    <body>
        <h1>🏍️ MotoGP 2025 Newsletter - Latest News 🏁</h1>
        <div class="newsletter-container">
            <h2>🔥 What is MotoGP? 🔥</h2>
            <p class="intro-text">
                MotoGP is the <b>pinnacle of motorcycle racing</b>, where <i>speed meets strategy</i>.
                This premier championship features <b>elite riders, high-tech bikes, and legendary circuits</b>.
                The <b>2025 season</b> brings <b>thrilling rivalries, emerging talents</b>, and <b>groundbreaking innovations</b> that push the limits of motorsport.
                Stay updated with our <b>exclusive race highlights, news, and insights!</b> 🚀
            </p>
            <h2>🏆 Latest News Updates 🏆</h2>
    """

    # Loop through articles and add them to HTML
    for _, news in news_df.iterrows():
        html_content += f"""
        <div class="news-box">
            <p class="news-title">{news['Title']}</p>
            <img src="{news['Image URL']}" alt="News Image" class="news-image">
            <p class="news-desc">{news['Description'][:400]}...</p>
            <a href="{news['Link']}" class="news-link" target="_blank">📖 Read Full Article</a>
            <p class="footer"><strong>Category:</strong> {news['Category']} | <strong>Published Date:</strong> {news['Published Date']}</p>
        </div>
        """

    html_content += """
        <p class="credits">Brought to you by Vatsal Nanawati 🚀</p>
        </div>
    </body>
    </html>
    """
    return html_content

# Fetch news data
news_data = fetch_motogp_news()

if news_data:
    news_df = extract_first_three_news(news_data)

    if not news_df.empty:
        display(news_df)
        newsletter_html = generate_newsletter_html(news_df)

        with open("motogp_newsletter.html", "w", encoding="utf-8") as file:
            file.write(newsletter_html)

        print("✅ Newsletter HTML file generated successfully!")
    else:
        print("⚠️ Not enough news articles found.")
else:
    print("❌ Failed to retrieve news data.")
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import os

# Gmail SMTP Server Configuration
SMTP_SERVER = "smtp.gmail.com"
SMTP_PORT = 587  # Use 465 for SSL or 587 for TLS

# Your Gmail Credentials (Use Environment Variables for Security)
EMAIL_ADDRESS = "nanawativatsal2000@gmail.com"  # Replace with your Gmail
EMAIL_PASSWORD = os.getenv("GMAIL_APP_PASSWORD")  # Use an environment variable for security

# Recipient Emails (Fix: Convert tuple to a list)
TO_EMAILS = [
        "akarrahe@ucdavis.edu"
]  # Ensure the last email is ".com", not ".co"

# Email Subject
SUBJECT = "🏍️ MotoGP Newsletter - Latest Updates 🏁"

# Read Newsletter HTML file
try:
    with open("motogp_newsletter.html", "r", encoding="utf-8") as file:
        html_content = file.read()
except FileNotFoundError:
    print("❌ Newsletter file not found! Ensure 'motogp_newsletter.html' is generated.")
    exit(1)

# Create Email Message
msg = MIMEMultipart()
msg["From"] = EMAIL_ADDRESS
msg["To"] = ", ".join(TO_EMAILS)  # Convert list to comma-separated string
msg["Subject"] = SUBJECT
msg.attach(MIMEText(html_content, "html"))  # Attach the HTML Newsletter

try:
    # Connect to Gmail SMTP Server
    server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
    server.starttls()  # Secure connection using TLS

    # Login using App Password
    server.login(EMAIL_ADDRESS, EMAIL_PASSWORD)

    # Send Email
    server.sendmail(EMAIL_ADDRESS, TO_EMAILS, msg.as_string())

    print("✅ Email sent successfully!")

    # Close Server
    server.quit()

except Exception as e:
    print(f"❌ Failed to send email: {e}")
```
