```python
!pip install geopandas
import geopandas as gpd

geojson_file = next(iter(geojson))  # Automatically grabs the uploaded file
gdf = gpd.read_file(geojson_file)      # Quick visualization (uses matplotlib)
```

```python
import folium
import geopandas as gpd
from folium import GeoJson

# Convert to GeoJSON-compatible format
geojson = gdf.to_json()

# Create interactive map centered around your data
center = [gdf.geometry.centroid.y.mean(), gdf.geometry.centroid.x.mean()]
m = folium.Map(location=center, zoom_start=10)

# Add GeoJSON layer
folium.GeoJson(geojson, name="geojson").add_to(m)

# Show map
m
```
